```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generar datos para la nube de puntos
np.random.seed(42)  # Para reproducibilidad
X1 = np.random.normal(loc=[2, 2], scale=0.8, size=(150, 2))
X2 = np.random.normal(loc=[6, 6], scale=0.8, size=(150, 2))
X = np.vstack([X1, X2])  # Concatenar las dos nubes de puntos

# Visualizacin inicial "Antes de k-means"
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c='black', s=10)
plt.title("Antes de k-means", fontsize=20)
plt.xlabel("X")
plt.ylabel("Y")

# Aplicar k-means con k=2
kmeans = KMeans(n_clusters=2, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Visualizacin "Despus de k-means"
plt.subplot(1, 2, 2)
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], c='blue', s=10, label='Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], c='orange', s=10, label='Cluster 2')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=100, marker='X', label='Centroids')
plt.title("Despus de k-means ($k=2$)", fontsize=20)
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Aplicar k-means para k=4
kmeans_k4 = KMeans(n_clusters=4, random_state=42)
y_kmeans_k4 = kmeans_k4.fit_predict(X)

# Aplicar k-means para k=5
kmeans_k5 = KMeans(n_clusters=5, random_state=42)
y_kmeans_k5 = kmeans_k5.fit_predict(X)

# Visualizacin para k=4 y k=5
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Visualizacin para k=4
for cluster in range(4):
    axes[0].scatter(X[y_kmeans_k4 == cluster, 0], X[y_kmeans_k4 == cluster, 1],
                    s=10, label=f'Cluster {cluster+1}')
axes[0].scatter(kmeans_k4.cluster_centers_[:, 0], kmeans_k4.cluster_centers_[:, 1],
                c='red', s=100, marker='X', label='Centroides')
axes[0].set_title("Despus de k-means ($k=4$)", fontsize=20)
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")
axes[0].legend()

# Visualizacin para k=5
for cluster in range(5):
    axes[1].scatter(X[y_kmeans_k5 == cluster, 0], X[y_kmeans_k5 == cluster, 1],
                    s=10, label=f'Cluster {cluster+1}')
axes[1].scatter(kmeans_k5.cluster_centers_[:, 0], kmeans_k5.cluster_centers_[:, 1],
                c='red', s=100, marker='X', label='Centroides')
axes[1].set_title("Despus de k-means ($k=5$)", fontsize=20)
axes[1].set_xlabel("X")
axes[1].set_ylabel("Y")
axes[1].legend()

plt.tight_layout()
plt.show()
```
